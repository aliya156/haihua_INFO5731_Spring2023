Question 2 (10 points): Write python code to extract these features you discussed above. You can collect a few sample text data for the feature extraction. 

```python
import nltk
nltk.download('all')
```

```python
# You code here (Please add comments in the code):
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import WordNetLemmatizer
import string

# Sample text data
text_data = "The acting was not good, but the cinematography was amazing. Overall, it was a fantastic movie!"

# Convert to lowercase
text_data = text_data.lower()

# Tokenize words
words = word_tokenize(text_data)

# Remove stopwords and punctuation
stopwords_list = stopwords.words('english') + list(string.punctuation)
words = [word for word in words if word not in stopwords_list]

# Lemmatize words
lemmatizer = WordNetLemmatizer()
words = [lemmatizer.lemmatize(word) for word in words]

# Word frequency feature
word_freq = {}
for word in words:
    if word not in word_freq:
        word_freq[word] = 1
    else:
        word_freq[word] += 1

# Parts of speech feature
pos_tags = nltk.pos_tag(words)
pos_counts = {}
for tag in pos_tags:
    if tag[1] not in pos_counts:
        pos_counts[tag[1]] = 1
    else:
        pos_counts[tag[1]] += 1

# Negation feature
negation_words = ['not', 'no', 'never', 'neither', 'nor', 'nobody', 'none']
has_negation = any(word in negation_words for word in words)

# Length of review feature
review_length = len(words)

# Punctuation feature
punctuation_counts = {}
for char in text_data:
    if char in string.punctuation:
        if char not in punctuation_counts:
            punctuation_counts[char] = 1
        else:
            punctuation_counts[char] += 1

print("Word frequency: ", word_freq)
print("Parts of speech: ", pos_counts)
print("Negation: ", has_negation)
print("Review length: ", review_length)
print("Punctuation counts: ", punctuation_counts)

```

```python
# You code here (Please add comments in the code):
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_selection import mutual_info_classif

# Sample text data
text_data = "The acting was not good, but the cinematography was amazing. Overall, it was a fantastic movie!"

# Convert to lowercase
text_data = text_data.lower()

# Tokenize words
words = word_tokenize(text_data)

# Remove stopwords and punctuation
stopwords_list = stopwords.words('english') + list(string.punctuation)
words = [word for word in words if word not in stopwords_list]

# Lemmatize words
lemmatizer = WordNetLemmatizer()
words = [lemmatizer.lemmatize(word) for word in words]

# Convert words to numeric values using CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform([' '.join(words)])
feature_names = vectorizer.get_feature_names()

# Calculate MI scores
y = ['positive']  # assuming this review is positive
mi_scores = mutual_info_classif(X, y)

# Print MI scores in descending order
sorted_features = sorted(zip(feature_names, mi_scores), key=lambda x: x[1], reverse=True)
for feature, score in sorted_features:
    print(f"{feature}: {score}")
```

```python
import numpy as np
import torch
from transformers import BertTokenizer, BertModel
from sklearn.metrics.pairwise import cosine_similarity

# Sample text data
text_data = ["This is a fantastic movie. I loved the acting and the cinematography.", 
             "The acting was not good, but the cinematography was amazing. Overall, it was a fantastic movie!",
             "I didn't like this movie at all. The plot was confusing and the acting was terrible."]

# Query to match the most relevant documents
query = "I enjoyed this movie. The acting was great and the cinematography was stunning."

# Load BERT model and tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
model = BertModel.from_pretrained('bert-base-uncased')

# Tokenize and encode query
query_tokens = tokenizer.tokenize(query)
query_tokens = ['[CLS]'] + query_tokens + ['[SEP]']
query_ids = tokenizer.convert_tokens_to_ids(query_tokens)
query_tensor = torch.tensor(query_ids).unsqueeze(0)

# Get BERT embeddings for query
with torch.no_grad():
    query_outputs = model(query_tensor)
    query_embeddings = query_outputs[0][:, 0, :].numpy().reshape(1, -1)

# Tokenize, encode and get BERT embeddings for text data
text_embeddings = np.zeros((len(text_data), 768))
for i, text in enumerate(text_data):
    text_tokens = tokenizer.tokenize(text)
    text_tokens = ['[CLS]'] + text_tokens + ['[SEP]']
    text_ids = tokenizer.convert_tokens_to_ids(text_tokens)
    text_tensor = torch.tensor(text_ids).unsqueeze(0)
    
    with torch.no_grad():
        text_outputs = model(text_tensor)
        text_embedding = text_outputs[0][:, 0, :].numpy()
        text_embeddings[i] = text_embedding

# Calculate cosine similarity between query and each text in data
similarities = cosine_similarity(query_embeddings, text_embeddings)[0]

# Rank the similarities in descending order
ranked_similarities = np.argsort(similarities)[::-1]
for i in ranked_similarities:
    print(f"Text {i+1}: {text_data[i]}")
    print(f"Cosine similarity: {similarities[i]}")
```
