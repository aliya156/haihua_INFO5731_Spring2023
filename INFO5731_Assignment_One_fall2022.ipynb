(20 points). Write a Python program to generate 12 random numbers between 1 and 100. Sort the list of numbers in ascending order and print it out. Sort the list of numbers in descending order and print it out. Calculate the average for the group and print it out.

```python
# Your code here
import random 
x=[]
for i in range(12):
    number = random.randrange(2,100)
    print(number)
    x.append(number)
x.sort()
print("Ascending order", x)
x.sort(reverse=True)
print("Decending order",x)
print("Average of list items", sum(x)/len(x))
```

# Question 2
(25 points). Write a program that will do the following string formatting tasks:

(1) Ask the user to enter two sentences.

(2) Make each sentence into a list. Each element in the list should correspond to a word in the sentence. For example, if the sentence is "Alas! Am I alive? This is my first python program", the program should print out ['Alas, '!', 'Am', 'I', 'alive, '?', 'This', 'is', 'my', 'first', 'python', 'program'].

(3) Compare the two sentences. Print out a list of words that occur in both sentences.

(4) Remove any punctuation from the sentence and print it back out. It should remove at least the following characters, but it can remove more: period(“.”), comma (“,”), semicolon (“;”), and colon (“:”)

(5) Count the number of vowels in the sentences. Print out each vowel and the number of times it appears in the sentences, such as: a:2, e:1, i:0, etc

```python
# Your code here
def sentence_to_list(sentence):
    sentence = sentence.replace(",", "").replace(".", "").replace(";", "").replace(":", "")
    sentence = sentence.split(" ")
    return sentence

def count_vowels(sentence):
    vowels = "aeiouAEIOU"
    vowel_count = {}
    for char in sentence:
        if char in vowels:
            if char in vowel_count:
                vowel_count[char] += 1
            else:
                vowel_count[char] = 1
    return vowel_count


sentence1 = input("Enter the first sentence: ")
sentence2 = input("Enter the second sentence: ")

sentence1 = sentence_to_list(sentence1)
sentence2 = sentence_to_list(sentence2)
print(sentence1)
print(sentence2)

common_words = [word for word in sentence1 if word in sentence2]
print("Words that occur in both sentences:", common_words)

sentence1_vowels = count_vowels("".join(sentence1))
sentence2_vowels = count_vowels("".join(sentence2))
combined_vowels = {vowel: sentence1_vowels.get(vowel, 0) + sentence2_vowels.get(vowel, 0) for vowel in set(sentence1_vowels) | set(sentence2_vowels)}
print("Vowel count:", combined_vowels)
```

Write a program to calculate the value of V when r is in the range of 1–10. Output the result in the following format:

```python
def harmonic_mean(values):

    reciprocals = [1/value for value in values]
    average_reciprocal = sum(reciprocals) / len(reciprocals)
    harmonic_mean = 1/average_reciprocal
    
    return harmonic_mean
```
